```python
import PySimpleGUI as sg

import smtplib 
from email.mime.multipart import MIMEMultipart 
from email.mime.text import MIMEText 
from email.mime.base import MIMEBase 
from email import encoders 

import certgen

import pandas

import os
```

```python
# string to store the body of the mail 
body = '''Hi!
Hope you're doing great!! 
Thanks for participating in the SEED3 Event.

Hereby attached is your certificate and a G-drive link for the e-kit.
Stay tuned for the session recordings :)

e-Kit: https://drive.google.com/drive/folders/1DHwDGvg5QX89bGQufxPb7urZ69Oqcnk2?usp=sharing

Please continue supporting us in our cause to provide quality education
May the Python, be with you!! 

--Team SEED
'''

def mailer(toaddr, filename, mailbody):
    
    fromaddr = "seed.raccbespectrum3201@gmail.com"
    
    # instance of MIMEMultipart 
    msg = MIMEMultipart() 
      
    # storing the senders email address   
    msg['From'] = fromaddr 
      
    # storing the receivers email address  
    msg['To'] = toaddr 
      
    # storing the subject  
    msg['Subject'] = "Python Level-1 Workshop"
      
    # attach the body with the msg instance 
    msg.attach(MIMEText(mailbody, 'plain')) 
      
    # open the file to be sent  
    attachment = open(filename, "rb") 
      
    # instance of MIMEBase and named as p 
    p = MIMEBase('application', 'octet-stream') 
      
    # To change the payload into encoded form 
    p.set_payload((attachment).read()) 
      
    # encode into base64 
    encoders.encode_base64(p) 
       
    p.add_header('Content-Disposition', "attachment; filename= %s" % filename) 
      
    # attach the instance 'p' to instance 'msg' 
    msg.attach(p) 
      
    # creates SMTP session 
    # Port 587 is the default mail submission port that supports all types of SMTP data transmission
    s = smtplib.SMTP('smtp.gmail.com', 587) 
      
    # start TLS for security 
    s.starttls() 
      
    # Authentication 
    s.login(fromaddr, "password") 
      
    # Converts the Multipart msg into a string 
    text = msg.as_string() 
      
    # sending the mail 
    s.sendmail(fromaddr, toaddr, text) 
      
    # terminating the session 
    s.quit() 
```

```python
sg.theme('Light Blue 2')

layout = [ [sg.Text('Img Path:', size=(8, 1)), sg.Input(key = '-imgpath-'), sg.FileBrowse("Other",size = (8,1))],
           [sg.Text('o/p Folder:', size=(8, 1)), sg.Input(key = '-destpath-'), sg.FolderBrowse(size = (8,1))],
           [sg.Text(' '*98),sg.Button('Gen Certs', key = '-gencert-',size = (8,1))],
           [sg.Text('e-mail Content:')],
           [sg.Multiline(default_text= body, size=(65, 10), key = '-body-')],
           [sg.Text('e-mail ids:', size=(8, 1)), sg.Input(key = '-csvpath-'), sg.FileBrowse(size = (8,1))],
           [sg.Text(' '*98),sg.Button('Send Certs', key = '-sendcert-',size = (8,1))]
         ]

window = sg.Window('SEED_Event Automation', layout)
```

```python

while True:
    event, values = window.read()
    
    if event == sg.WIN_CLOSED:
        break
        
    if(event == '-sendcert-'):
        
        csvpath = values['-csvpath-']
        destpath = values['-destpath-']
        
        data = pandas.read_csv(csvpath, names = ['id','names','mail']) 
        
        mailtxt = data.mail.tolist()
        idtxt = data.id.tolist() 
        
        for ids in range(1, len(mailtxt)):
            mailer(mailtxt[ids], os.path.join(os.path.normpath(destpath), idtxt[ids]+".jpg"), values['-body-'])
            print(mailtxt[ids] + " Sent!")
            
    if(event == '-gencert-'):

        imgpath = values['-imgpath-']
        csvpath = values['-csvpath-']
        destpath = values['-destpath-']

        data = pandas.read_csv(csvpath, names = ['id','names','mail'])
   
        idtxt = data.id.tolist() 
        nametxt = data.names.tolist()
        
        
        print(imgpath,csvpath,destpath)
        
        certgen.certgen(imgpath, csvpath, destpath)
        
        print("Hi, your answer!")

  
window.close()
```
